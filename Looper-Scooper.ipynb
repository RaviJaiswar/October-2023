```python
import numpy as np
from numpy import *
from numpy import nan
import os
import astropy
import pandas as pd
from astropy.io import fits
from astropy.table import Table
from astropy.visualization import astropy_mpl_style
from astropy.utils.data import get_pkg_data_filename
from astropy.cosmology import WMAP9 as cosmo
from astropy.cosmology import FlatLambdaCDM
import astropy.units as u
```

```python
path = os.path.expanduser('~/Documents/mag2/')
Montague = pd.read_csv(os.path.expanduser(path+'files/observations.dat')) 
Parameters=[r'SFH $f_{mu}$',r'IR $f_{mu}$',r'$\mu$',r'$\tau_V$',r'sSFR$_{0.1Gyr}$','M(stars)','Ldust',r'$T_C^{ISM}$',r'$T_W^{BC}$',r'$\chi_C^{tot}$',r'$\chi_{PAH}^{tot}$',r'$\chi_{MIR}^{tot}$',r'$\chi_{W}^{tot}$',r'$\tau_{V}^{ISM}$','M(dust)',r'$SFR_{0.1Gyr}$']
Cheese=[]
Toast=[]
runawayfaster=[]
for Curl in range(len(Montague)):
    potato=str(Montague['id'][Curl])
#     potato=str(735)
    Minestrone = pd.read_csv(os.path.expanduser(path+'/'+potato+'.fit'),index_col=0,on_bad_lines='skip')
    Minestrone2 = pd.DataFrame(columns=['Full'])
    potato=int(potato)
    redshift =float(Minestrone.index[6].split()[-1])
    runawayfaster.append(redshift)
    line = Minestrone.index[8].split(" ")
    while("" in line):
        line.remove("")
    str_line = " ".join(str(item) for item in line)
    Minestrone2.loc[0]=str(str_line)
    Minestrone2[Parameters] = Minestrone2['Full'].str.split(" ",expand=True,)
    Minestrone2=Minestrone2.drop(['Full'],axis=1)
    Cheese.append(Minestrone2.iloc[0])
    Toast.append(potato)
```

```python
Carpenter=Table(Cheese).to_pandas()

Carpenter=transpose(Carpenter).reset_index(drop=True)

Carpenter.columns = [Parameters]

Carpenter.insert(0,'Galaxy_ID',Toast)

Carpenter.insert(1,'redshift',runawayfaster)
```
