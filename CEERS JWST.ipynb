```python
def check_matching_coordinates(input_ra, input_dec, table):
    # Extract the RA and DEC columns from the table
    ra = table['ra']
    dec = table['dec']

    tolerance = 0.0002777*2# degrees - 0.0002777 = 1arcsec
    match_indices = []

    for i in range(len(ra)):
        sep = np.sqrt((input_ra - ra[i])**2 + (input_dec - dec[i])**2)

        if sep < tolerance:
            match_indices.append(i)

    return match_indices
check_matching_coordinates(ra_deg,dec_deg,x)
```
